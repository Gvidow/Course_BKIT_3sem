### Построение графика

```python
url = "http://localhost:8080/catalan/"
def get_data(num):
    return requests.get(url + str(num)).json()
```

```python
y_10 = get_data(10)
x_10 = [i for i in range(1, 11)]
```

##### Ось абсцисс

```python
x_10
```

##### Ось ординат

```python
y_10
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(x_10, y_10)
plt.xlabel('порядковый номер')
plt.ylabel('Число Каталана')
plt.title('Первые {} чисел последовательности Каталана'.format(len(y_10)))
plt.show()
```
